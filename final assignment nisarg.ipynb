```python
Y = dist.iloc[17:,1].values
Z = dist.iloc[17:,2].values
R = dist.iloc[17:,3].values
D = dist.iloc[17:,4].values
X = dist.iloc[17:,0]

plt.plot(X, Y)

##here values are assigned as Y,Z etc...
```

```python
plt.plot(X, Y, label="active cases", color="b", linewidth=2, marker='o',markersize=7)
plt.xlabel('district')
plt.ylabel('No. of cases')
plt.title('COVID ANALYSIS')
plt.legend() 
plt.show() 
```

```python
plt.plot(X, Z, label="confirmed cases", color="r", linewidth=2, marker='+',markersize=7)
plt.xlabel('district')
plt.ylabel('No. of cases')
plt.title('COVID ANALYSIS')
plt.legend() 
plt.show() 
```

```python
plt.plot(X, R, label="deceased cases", color="g", linewidth=2, marker='*',markersize=12)
plt.xlabel('district')
plt.ylabel('No. of cases')
plt.title('COVID ANALYSIS')
plt.legend() 
plt.show() 
```

```python
plt.plot(X, D, label="recovered cases", color="black", linewidth=2, marker='s',markersize=8)
plt.xlabel('district')
plt.ylabel('No. of cases')
plt.title('COVID ANALYSIS')
plt.legend() 
plt.show() 
```

```python
##GRAPH HAVING ACTIVE CASES VS CONFIRMED VS DECEASED VS RECOVERED
plt.plot(X, Y, label="active cases", color="b", linewidth=2, marker='o',markersize=7)
plt.plot(X, Z, label="confirmed cases", color="r", linewidth=2, marker='+',markersize=7)
plt.plot(X, R, label="deceased cases", color="g", linewidth=2, marker='*',markersize=12)
plt.plot(X, D, label="recovered cases", color="black", linewidth=2, marker='s',markersize=8)

plt.xlabel('district')
plt.ylabel('No. of cases')
plt.title('COVID ANALYSIS')
plt.legend() 
plt.show() 
```
